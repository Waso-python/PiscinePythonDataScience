```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_json('../ex02/auto.json')
```

```python
pd.options.display.float_format = '{:.2f}'.format
df
```

```python
temp = df.sample(n=200, random_state=21)
temp
```

```python
concat_rows = pd.concat([df, temp], ignore_index=True)
concat_rows
```

```python
np.random.seed(21)
concat_rows['Year'] = np.random.randint(1980, 2019, size=concat_rows['CarNumber'].count())
fines = concat_rows.copy()
fines
```

```python
surname = pd.read_json('../data/surname.json')
surname
```

```python
surname_serial = surname.sample(n=fines.CarNumber.drop_duplicates().count(), random_state=21, replace=True, ignore_index=True)
surname_serial
```

```python
owners = pd.DataFrame()
owners['CarNumber'] = fines.CarNumber
owners.drop_duplicates(['CarNumber'], inplace=True, ignore_index=True)
owners['SURNAME'] = surname_serial[0]
owners
```

```python
# temp = fines.copy()
temp = pd.DataFrame({'CarNumber':['PD163O3161RUS', 'PD963O8161RUS', 'PD163O4161RUS', 'PD163O8431RUS', 'PD16398161RUS'],
                     'Refund':[1,2,3,2,1],
                     'Fines':[3700,2900,2350,2650,4500],
                     'Make':['Porshe', 'Mitsubishi', 'KIA', 'Triumph', 'Jiga'],
                     'Model':['Panamera', 'Lancer', 'Rio', 'Spitfire', '8'],
                     'Year':[2001,2002,2003,2004,2005]})

fines = pd.concat([fines, temp], ignore_index=True)
fines
```

```python
owners.drop(owners.tail(20).index, inplace=True)
temp = pd.DataFrame({'CarNumber':['Y163O8431GG','Y163O8231GG', 'Y163O9431GGG'],
                     'SURNAME':['Ivanov', 'Petrov', 'Sidorov']})

owners = pd.concat([owners, temp], ignore_index=True)
owners
```

```python
new1 = pd.merge(fines, owners, on='CarNumber', how='inner')
new1
```

```python
new2 = pd.merge(fines, owners, on='CarNumber', how='outer')
new2
```

```python
new3 = pd.merge(fines, owners, on='CarNumber', how='left')
new3
```

```python
new4 = pd.merge(fines, owners, on='CarNumber', how='right')
new4
```

```python
pd.pivot_table(fines, columns='Year', values='Fines', index=['Make', 'Model'], aggfunc={'Fines': np.sum})
```

```python
fines.to_csv('fines.csv')
owners.to_csv('owners.csv')
fines
```
